```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Piloto': ['Max Verstappen', 'Lewis Hamilton', 'Charles Leclerc', 'Sergio Perez', 'George Russell'],
    'Puntos': [575, 234, 206, 285, 175],
    'Victorias': [19, 0, 0, 2, 0],
    'Podios': [21, 6, 7, 9, 7],
    'Pole_Positions': [12, 1, 5, 2, 0]
}
df = pd.DataFrame(data)
fig, gc = plt.subplots(figsize=(10, 6))

gc.plot(df['Piloto'], df['Puntos'], label='Puntos Totales', color='blue', marker='o', linestyle='-')
gc.set_xlabel('Piloto')
gc.set_ylabel('Puntos (Eje Izquierdo)', color='blue')
gc.grid(True, linestyle='-', axis='y')
gc.tick_params(axis='x', rotation=45)

gc2 = gc.twinx()
gc2.plot(df['Piloto'], df['Podios'], label='NÃºmero de Podios', color='red', marker='s', linestyle='.')
gc2.set_ylabel('Podios', color='red')
gc2.tick_params(axis='y', labelcolor='red')

fig.suptitle('Puntos y Podios por Piloto de F1')
fig.legend(loc='upper right', bbox_to_anchor=(0.9, 0.9))
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))

plt.plot(df['Piloto'], df['Victorias'], label='Victorias', color='green', marker='^')
plt.plot(df['Piloto'], df['Pole_Positions'], label='Pole Positions', color='purple', marker='d', linestyle=':')

plt.title('Victorias vs Pole Positions por Piloto')
plt.xlabel('Piloto')
plt.ylabel('Conteo')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
