```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "Piloto": ["Max Verstappen", "Lewis Hamilton", "Charles Leclerc", "Sergio Perez", "George Russell"],
    "Puntos": [575, 234, 206, 285, 175],
    'Victorias': [19, 0, 0, 2, 0],
    "Podios": [21, 6, 7, 9, 7],
    "Pole_Positions": [12, 1, 5, 2, 0]
}
df = pd.DataFrame(data)
fig, gc = plt.subplots(figsize=(10, 6))

gc.plot(df['Piloto'], df['Puntos'], label='Puntos Totales', color='blue', marker='o', linestyle='-')
gc.set_xlabel('Piloto')
gc.set_ylabel('Puntos (Eje Izquierdo)', color='blue')
gc.grid(True, linestyle='-', axis='y')
gc.tick_params(axis='x', rotation=45)

gc2 = gc.twinx()
gc2.plot(df['Piloto'], df['Podios'], label='Número de Podios', color='red', marker='s', linestyle='.')
gc2.set_ylabel('Podios', color='red')
gc2.tick_params(axis='y', labelcolor='red')

fig.suptitle('Puntos y Podios por Piloto de F1')
fig.legend(loc='upper right', bbox_to_anchor=(0.9, 0.9))
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))

plt.plot(df['Piloto'], df['Victorias'], label='Victorias', color='green', marker='^')
plt.plot(df['Piloto'], df['Pole_Positions'], label='Pole Positions', color='purple', marker='d', linestyle=':')

plt.title('Victorias vs Pole Positions por Piloto')
plt.xlabel('Piloto')
plt.ylabel('Conteo')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#Tarea 4 tipos de grafica, clase 14

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    "Piloto": ["Max Verstappen", "Lewis Hamilton", "Charles Leclerc", "Sergio Perez", "George Russell"],
    "Puntos": [575, 234, 206, 285, 175],
    "Victorias": [19, 0, 0, 2, 0],
    "Podios" : [21, 6, 7, 9, 7],
    "Pole_Positions ": [12, 1, 5, 2, 0]
}
df = pd.DataFrame(data)

fig, axes = plt.subplots(2, 2, figsize=(15, 12))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# Gráfico de barras
ax1 = axes[0, 0]
colores_pilotos = ['blue', 'silver', 'red', 'darkblue', 'gray']
ax1.bar(df['Piloto'], df['Victorias'], color=colores_pilotos)
ax1.set_title('Gráfico de Barras: Victorias por Piloto')
ax1.set_xlabel('Piloto')
ax1.set_ylabel('Número de Victorias')
ax1.grid(axis='y', alpha=0.5)
ax1.tick_params(axis='x', rotation=20)

# Gráfico de Pastel
ax2 = axes[0, 1]
ax2.pie(
    df['Puntos'],
    labels=df['Piloto'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colores_pilotos
)
ax2.set_title('Gráfico de Pastel: Proporción de Puntos Totales')
ax2.axis('equal')

# Grafico de Scatter
# Grafico de Scatter
ax3 = axes[1, 0]
ax3.scatter(df["Puntos"], df["Podios"], color='green', marker='o', s=180, alpha=0.8)
ax3.set_title("Scatter: Puntos vs. Podios")
ax3.set_xlabel('Puntos Totales')
ax3.set_ylabel('Número de Podios')

for i, piloto in enumerate(df['Piloto']):
    ax3.annotate(piloto, (df['Puntos'][i] + 5, df['Podios'][i] - 0.2), fontsize=5)
ax3.grid(axis='both', alpha=0.3)
offsets = {
    "Max Verstappen": (10, -3),
    "Sergio Perez": (9, -4),
    "Lewis Hamilton": (9, -5),
    "Charles Leclerc": (10, -2),
    "George Russell": (5, 10)
}

# Grafico Histogramas
ax4 = axes[1, 1]
ax4.hist(df['Puntos'], bins=5, color='purple', edgecolor='black', alpha=0.7)
ax4.set_title( "Gráfico de Histogramar: Distribución de Puntos")
ax4.set_xlabel("Rangos de Puntos")
ax4.set_ylabel('Frecuencia (Número de Pilotos)')
ax4.grid(axis="y", alpha=0.5)

plt.show()
```

#Clase 15 Seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
num_races = 22
data = []
years = [2021, 2022, 2023]

for year in years:
    points_base = 25 if year in [2022, 2023] else 20
    points = np.clip(np.random.normal(points_base, 5, num_races), 0, 26)
    position_base = 3 if year in [2022, 2023] else 5
    position = np.clip(np.random.normal(position_base, 3, num_races), 1, 20).astype(int)
    pit_time = np.clip(np.random.normal(2.5, 0.5, num_races), 1.8, 4.0)
    speed = np.clip(np.random.normal(330, 5, num_races), 320, 340)

    for i in range(num_races):
        data.append({
            'Race': f'R{i+1}',
            'Year': year,
            'Points': points[i],
            'Position': position[i],
            'Pit_Time_s': pit_time[i],
            'Top_Speed_kmh': speed[i],
            'Driver': 'Max Verstappen'
        })

df = pd.DataFrame(data)

sns.set_style('whitegrid')
sns.set_context('talk')

fig, axes = plt.subplots(2, 3, figsize=(24, 16))
fig.suptitle("Rendimiento de Max Verstappen (2021-2023)", fontsize=30, y=1.02)

sns.boxplot(data=df, x='Year', y='Points', ax=axes[0, 0])
axes[0, 0].set_title('1. Distribución de Puntos por Carrera (Boxplot)')

sns.violinplot(data=df, x='Year', y='Position', ax=axes[0, 1])
axes[0, 1].set_title('2. Densidad de Posiciones Finales (Violinplot)')
axes[0, 1].invert_yaxis()

sns.scatterplot(data=df, x='Position', y='Points', hue='Year', style='Year', palette='rocket', s=200, ax=axes[0, 2])
axes[0, 2].set_title('3. Relación Puntos vs Posición Final')
axes[0, 2].invert_xaxis()

sns.set_style('darkgrid')
sns.histplot(data=df, x='Pit_Time_s', hue='Year', kde=True, ax=axes[1, 0])
axes[1, 0].set_title('4. Distribución de Tiempos de Pit Stop')

corr_matrix = df[['Points', 'Position', 'Pit_Time_s', 'Top_Speed_kmh']].corr()

sns.heatmap(corr_matrix,
            annot=True,
            cmap='viridis',
            fmt=".2f",
            linewidths=.5,
            cbar_kws={'label': 'Coeficiente de Correlación'},
            ax=axes[1, 1])
axes[1, 1].set_title('5. Heatmap de Correlación entre Variables')

sns.set_style('whitegrid')
sns.lineplot(data=df, x='Race', y='Top_Speed_kmh', hue='Year', marker='o', palette='crest', ax=axes[1, 2])
axes[1, 2].set_title('6. Velocidad Punta por Carrera y Año')
axes[1, 2].set_xticklabels(df['Race'].unique(), rotation=90)

plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```
