```python
def sistema_estudiantes():
    estudiantes = [
        {"nombre": "Ana", "edad": 16, "grado": "11A", "materias": ["Arte", "Matemáticas"]},
        {"nombre": "Luis", "edad": 17, "grado": "12B", "materias": ["Física", "Historia"]},
        {"nombre": "María", "edad": 15, "grado": "10C", "materias": ["Inglés", "Química"]},
        {"nombre": "Pedro", "edad": 16, "grado": "11A", "materias": ["Biología", "Geografía"]},
        {"nombre": "Sofía", "edad": 17, "grado": "12A", "materias": ["Música", "Literatura"]}
    ]

    while True:
        print("\n=== SISTEMA DE ESTUDIANTES ===")
        print("1. Ver lista")
        print("2. Agregar estudiante")
        print("3. Eliminar estudiante")
        print("4. Actualizar estudiante")
        print("5. Salir")
        op = input("Elige una opción: ")

        if op == "1":
            for e in estudiantes:
                print(f"Nombre: {e['nombre']} | Edad: {e['edad']} | Grado: {e['grado']} | Materias: {', '.join(e['materias'])}")

        elif op == "2":
            n = input("Nombre: ")
            e = int(input("Edad: "))
            g = input("Grado: ")
            m = input("Materias favoritas (separa con comas): ").split(",")
            estudiantes.append({"nombre": n, "edad": e, "grado": g, "materias": [x.strip() for x in m]})
            print("Estudiante agregado")

        elif op == "3":
            n = input("Nombre del estudiante a eliminar: ")
            for est in estudiantes:
                if est["nombre"].lower() == n.lower():
                    estudiantes.remove(est)
                    print("Estudiante eliminado")
                    break
            else:
                print("No encontrado")

        elif op == "4":
            n = input("Nombre del estudiante a actualizar: ")
            for est in estudiantes:
                if est["nombre"].lower() == n.lower():
                    est["nombre"] = input("Nuevo nombre (deja vacío si no cambia): ") or est["nombre"]
                    nueva_edad = input("Nueva edad: ")
                    if nueva_edad:
                        est["edad"] = int(nueva_edad)
                    est["grado"] = input("Nuevo grado: ") or est["grado"]
                    nuevas_m = input("Nuevas materias (separa con comas): ")
                    if nuevas_m:
                        est["materias"] = [x.strip() for x in nuevas_m.split(",")]
                    print("Información actualizada")
                    break
            else:
                print("No encontrado")

        elif op == "5":
            print("Gracias por usar el sistema")
            break
        else:
            print("Opción inválida ")

sistema_estudiantes()
```
